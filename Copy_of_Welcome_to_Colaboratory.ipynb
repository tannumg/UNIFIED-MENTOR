```
# prompt: remove duplicates from data

data = data.drop_duplicates()
print(data)
```

```
Year=['2000-01', '2001-02', '2002-03', '2003-04', '2004-05',
       '2005-06', '2006-07', '2007-08', '2008-09', '2009-10', '2010-11',
       '2011-12', '2012-13', '2013-14', '2014-15', '2015-16', '2016-17']
Sector=['Sector']
```

```
data.info()
```

```
#wide to long
melt=pd.melt(data,id_vars=Sector,value_vars=Year)
melt
```

```
sorted=melt.sort_values(['Sector','variable'],ignore_index=True)
sorted
```

```
print("STATS for SECTORS")
print(pd.DataFrame(sorted.groupby('Sector').describe().loc[:,:]).T)
```

```
sectorwise=sorted.groupby('Sector').sum()
sectorwise
```

```
sectorwise.plot(kind='bar',y='value')
```

```
sectorwise_fdi=data[['Sector','FDI','USD']]
print(sectorwise_fdi.head())
```

```
merged=pd.merge(sectorwise,sectorwise_fdi,on='Sector')
merged
```
